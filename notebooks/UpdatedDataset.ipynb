```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/My Drive/ISYE 3030 Project/FinalDataset.csv')

number_columns = ["Number of People in Poverty",
                   "Number of Unemployed People", "Housing Affordability (Median Home Value)",
                   "Number of People with Health Insurance Coverage", "Number of People With Educational Attainment (HS+)",
                   "Number of People With Internet Access"]
for col in number_columns:
  new_col = col.replace("Number of", "Percentage of")
  df[new_col] = df[col] / df["Final Total Population"] #Dividing this number columns by the estimated Total Population data.
  df.drop(columns=col, inplace= True)
df.to_csv('/content/drive/My Drive/ISYE 3030 Project/Regression_Dataset.csv', index=False)

```
